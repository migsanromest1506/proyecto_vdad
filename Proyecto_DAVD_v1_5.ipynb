# Agrupacion de datos y graficas

## cantidad de municipios que NO cuentan servicio de 2G en el primer tercio de 2022 y el primer tercio de 2023

```python
consulta= cobertura.query('COBERT_2G == "N" and ANO == 2022 and TRIMESTRE == 1')
consulta


```

```python
consulta1_1= cobertura.query('COBERT_2G == "N" and ANO == 2023 and TRIMESTRE == 1')
consulta1_1
```

## Agrupacion por proveedores para ver el total

```python
consultas= consulta.groupby('PROVEEDOR').count()
consultas
```

```python
consultas1= consulta1_1.groupby('PROVEEDOR').count()
consultas1
```

### Grafico del 2022

```python
consulta.groupby('PROVEEDOR').size().plot(kind='barh', color=sns.palettes.mpl_palette('Dark2'))
plt.gca().spines[['top', 'right',]].set_visible(False)
```

### Grafico del 2023

```python
consulta1_1.groupby('PROVEEDOR').size().plot(kind='barh', color=sns.palettes.mpl_palette('Dark2'))
plt.gca().spines[['top', 'right',]].set_visible(False)
```
